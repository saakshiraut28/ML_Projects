## Transforming data using map fuction.

### Map fuction is used to map the data.

```python
Raw_Housing_Data['No of Times Visited'].unique()
```

```python
mapping = {
    'None' : 0,
    'Once' : 1,
    'Twice' : 2,
    'Thrice' : 3,
    'Four' : 4,
}

Raw_Housing_Data['No of Times Visited']=Raw_Housing_Data['No of Times Visited'].map(mapping)
```

```python
Raw_Housing_Data['No of Times Visited'].unique()
```

```python
Raw_Housing_Data['Renovated Year']
```

## Where Function

### where function is used to set a condition. A function in numpy module.

```python
Raw_Housing_Data['Ever Renovated']=np.where(Raw_Housing_Data['Renovated Year']==0, 'No', 'Yes')
```

```python
Raw_Housing_Data
```

## Datetime Index function

### Function from pandas library.

```python
Raw_Housing_Data['Purchase Year'] = pd.DatetimeIndex(Raw_Housing_Data['Date House was Sold']).year
```

```python
Raw_Housing_Data
```

```python

Raw_Housing_Data["Years since renovation"] = np.where(Raw_Housing_Data['Ever Renovated'] == 'No', 0, 
                                                      Raw_Housing_Data['Purchase Year']-Raw_Housing_Data['Renovated Year'])
```

```python
Raw_Housing_Data
```

## To delete columns

### inplace : Make changes int the actual files.

```python
Raw_Housing_Data.drop(columns = ['Date House was Sold', 'Purchase Year','Renovated Year'], inplace=True)
```
