### FEATURE SCALING

```python
# Methods of feature scaling:

# Min-Max -> Value of range of transformed variable lies between [1,0]
# - Formula : x = (currentval - min(currentval))/(max(currentval)-min(currentval))


# Standardisation -> Rescales the feature value so that it has STANDARD NORMAL DISTRIBUTION with mean = 0 
#                    and Standard Deviation = 1
# - Formula : x = (currentvalue - mean / standard deviation

# Normalization -> The values is always fixed from -1 to 1
# - Formula : x = (currentval - mean(currentval))/(max(currentval)-min(currentval))
```
